```python
fileid_year = dict()

for id_name in os.listdir("/home/nedo/data/edinet/unzip"):
    fileid_year[id_name] = {}
    input_dir = os.path.join("/home/okajima/edinet/edinet/ma/data/output/mna/",id_name)

   
    for json_file in os.listdir(input_dir):

        with open(input_dir+'/'+json_file) as f:
            all_objs = json.load(f)
            for i in all_objs:
                try: 
                    if i["attributes"]["name"] == "jpdei_cor:CurrentFiscalYearStartDateDEI":
                        fileid_year[id_name]['currentStart'] = i['value']

                    if i["attributes"]["name"] == "jpdei_cor:CurrentPeriodEndDateDEI":
                        fileid_year[id_name]['currentEnd'] = i['value']                       
                except:
                    continue
```

```python
# store to disk

with open('./output/fileid_year', 'w') as f:
    json.dump(fileid_year, f)

```
